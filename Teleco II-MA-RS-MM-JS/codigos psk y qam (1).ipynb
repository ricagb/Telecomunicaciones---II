```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
N = 4  # Número de símbolos
M = 4  # Número de fases (#PSK)
T = 1  # Duración de cada símbolo
oversampling_factor = 10  # Factor de oversampling

# Generar símbolos aleatorios
bits = np.random.randint(M, size=N)

# Mapeo de símbolos a fases
phases = 2 * np.pi * bits / M
phases2 = np.linspace(0, 2 * np.pi, M, endpoint=False)
symbols = np.exp(1j * phases2)

# Desfase
symbols_desfase = np.exp(1j * (phases2 + np.pi / 4))

# Generar señal modulada
t = np.linspace(0, N * T, N * 100)
t2 = np.linspace(0, N, N * oversampling_factor)
signal = np.zeros(len(t))
for i in range(N):
    signal[i * 100: (i + 1) * 100] = np.cos(2 * np.pi * t[i * 100: (i + 1) * 100] / T + phases[i])

# Agregar ruido a la señal
SNR_dB = 10  # Relación señal-ruido en dB
SNR = 10 ** (SNR_dB / 10)
noise_power = 1 / SNR
noise = np.random.normal(0, np.sqrt(noise_power), len(t))
received_signal = signal + noise

# Demodulación
demodulated_bits = np.zeros(N, dtype=int)
for i in range(N):
    symbol = np.exp(1j * phases[i])  # Símbolo de referencia
    distances = np.abs(received_signal[i * 100: (i + 1) * 100] - symbol)
    demodulated_bits[i] = np.argmin(distances)

# Gráficas
plt.figure(figsize=(15, 10))

# Gráfico de la señal de salida
plt.figure(figsize=(8, 6))
colors = plt.cm.viridis(np.linspace(0, 1, M))  # Colores para cada símbolo
for i in range(N):
    plt.plot(t[i * 100: (i + 1) * 100], signal[i * 100: (i + 1) * 100], color=colors[bits[i]], alpha=0.5)
plt.title('Señal de Salida')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Graficar constelación
plt.figure(figsize=(6, 6))
plt.scatter(np.real(symbols_desfase), np.imag(symbols_desfase), color='b')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.title('Diagrama de Constelaciones')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.grid(False)

# Agregar etiquetas
labels = ['11', '10', '00', '01']
for i, label in enumerate(labels):
    plt.text(np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), label, ha='right', va='bottom')

plt.show()

# Graficar diagrama fasorial
plt.figure(figsize=(6, 6))
for i in range(M):
    plt.arrow(0, 0, np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), color='b',
              head_width=0.1, head_length=0.1, length_includes_head=True)
    angle_rad = np.angle(symbols_desfase[i])
    angle_deg = np.degrees(angle_rad)
    if angle_deg > 180:
        angle_deg -= 360
    if angle_deg > 0:
        plt.text(np.real(symbols_desfase[i]) / 2, np.imag(symbols_desfase[i]) / 2,
                 f'{angle_deg:.1f}°', ha='center', va='center')
    else:
        plt.text(np.real(symbols_desfase[i]) / 2, np.imag(symbols_desfase[i]) / 2,
                 f'{angle_deg:.1f}°', ha='center', va='center')

plt.plot([0], [0], 'ro', linewidth=1)  # Punto de referencia en el origen
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.title('Diagrama Fasorial')
plt.xlabel('Eje Real')
plt.ylabel('Eje Imaginario')
plt.grid(False)
plt.axis('equal')

# Agregar etiquetas
for i, label in enumerate(labels):
    plt.text(np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), label, ha='right', va='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
N = 4  # Número de símbolos
M = 4  # Número de fases (8QAM)
T = 1  # Duración de cada símbolo
oversampling_factor = 100  # Factor de oversampling

# Generar símbolos aleatorios
bits = np.random.randint(M, size=N)

# Mapeo de símbolos a fases
phases2 = np.linspace(0, 2 * np.pi, M, endpoint=False)
symbols_desfase = np.exp(1j * (phases2 + np.pi / 4))

# Generar otros cuatro puntos con magnitud menor pero en la misma posición y ángulo
symbols_desfase_menor = symbols_desfase * 0.5  # Magnitud reducida a la mitad

# Graficar constelación
plt.figure(figsize=(6, 6))
plt.scatter(np.real(symbols_desfase), np.imag(symbols_desfase), color='b', marker='o')  # Puntos azules
plt.scatter(np.real(symbols_desfase_menor), np.imag(symbols_desfase_menor), color='r', marker='o')  # Puntos rojos
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.title('Diagrama de Constelaciones')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.grid(False)

# Agregar etiquetas
labels = ['111', '101', '001', '011', '110', '100', '000', '010']  # Etiquetas de los puntos
for i, label in enumerate(labels):
    if i < 4:
        plt.text(np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), label, ha='right', va='bottom')
    else:
        plt.text(np.real(symbols_desfase_menor[i - 4]), np.imag(symbols_desfase_menor[i - 4]), label, ha='right', va='bottom')

plt.show()

# Graficar diagrama fasorial
plt.figure(figsize=(6, 6))
for i in range(M):
    plt.arrow(0, 0, np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), color='b',
              head_width=0.1, head_length=0.1, length_includes_head=True)
    plt.arrow(0, 0, np.real(symbols_desfase_menor[i]), np.imag(symbols_desfase_menor[i]), color='r',
              head_width=0.1, head_length=0.1, length_includes_head=True)  # Flechas adicionales
    angle_rad = np.angle(symbols_desfase[i])
    angle_deg = np.degrees(angle_rad)
    if angle_deg > 180:
        angle_deg -= 360
    if angle_deg > 0:
        plt.text(np.real(symbols_desfase[i]) / 2, np.imag(symbols_desfase[i]) / 2,
                 f'{angle_deg:.1f}°', ha='center', va='center')
    else:
        plt.text(np.real(symbols_desfase[i]) / 2, np.imag(symbols_desfase[i]) / 2,
                 f'{angle_deg:.1f}°', ha='center', va='center')

plt.plot([0], [0], 'ro', linewidth=1)  # Punto de referencia en el origen
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.title('Diagrama Fasorial')
plt.xlabel('Eje Real')
plt.ylabel('Eje Imaginario')
plt.grid(False)
plt.axis('equal')

# Agregar etiquetas
for i, label in enumerate(labels):
    if i < 4:
        plt.text(np.real(symbols_desfase[i]), np.imag(symbols_desfase[i]), label, ha='right', va='bottom')
    else:
        plt.text(np.real(symbols_desfase_menor[i - 4]), np.imag(symbols_desfase_menor[i - 4]), label, ha='right', va='bottom')

plt.show()

# Generar señales moduladas para cada uno de los 8 símbolos con las amplitudes correspondientes
t = np.linspace(0, T, T * oversampling_factor, endpoint=False)
colors = plt.cm.viridis(np.linspace(0, 1, 8))  # Colores para cada símbolo

amplitudes = [1.8, 1.8, 1.8, 1.8, 0.71, 0.71, 0.71, 0.71]  # Amplitudes correspondientes

plt.figure(figsize=(15, 8))
for i in range(4):
    signal = amplitudes[i] * np.cos(2 * np.pi * t + np.angle(symbols_desfase[i]))
    plt.plot(t + i*T, signal, color=colors[i], label=f' {labels[i]}')

for i in range(4):
    signal = amplitudes[i + 4] * np.cos(2 * np.pi * t + np.angle(symbols_desfase_menor[i]))
    plt.plot(t + (i + 4)*T, signal, color=colors[i + 4], label=f'{labels[i + 4]}')

plt.title('Señales Moduladas de Salida')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
